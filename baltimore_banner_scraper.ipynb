```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup
import json
```

```python
# get_request = requests.get("https://www.thebaltimorebanner.com/community/criminal-justice/road-rage-ZF3BRQMVVVELFPYVHOGYZSD2LI/")
get_request = requests.get("https://www.thebaltimorebanner.com/community/transportation/amtrak-job-cuts-baltimore-tunnel-JORG4WVRO5GOLOU7M7HDG4SVUU/")

print(get_request.status_code)
#request's content in bytes (huh?)
# print(get_request.content)

#request's content in text
print(get_request.text)
# print(get_request.headers)
```

```python
# soup = BeautifulSoup(get_request.text, "html.parser")
#according to tutorial, should pass in content instead of text to avoid problems with character encoding
soup = BeautifulSoup(get_request.content, "html.parser")

#but when I do the line above, I get ‚Äô replacing '
print(soup)
```

```python
#length of article headline
headline = soup.find("strong", class_ = "headline__bold")
print(headline)

headline_text = headline.text
print(headline_text)

length_headline_text_characters = len(headline_text)

headline_words = headline_text.split()
print(headline_words)
length_headline_text_words = len(headline_words)

print("length of article title is "+ str(length_headline_text_characters) + " characters and " + str(length_headline_text_words) + " words")
```

```python
# length of article body

body = soup.find("div", class_ = "article-body")

print(body)

body_text = body.text
print(body_text)

body_words = body_text.split()
print(body_words)
length_body_text_words = len(body_words)

print("length of article body is " + str(length_body_text_words) + " words")
```

```python
#date posted

date = soup.find("span", class_ = "attribution-date__date__published")

date_text = date.text
print(date_text)
```

```python
#category?

category = soup.find("a", class_ = "default-variant__section-front")

category_text = category.text
print(category_text)
```

```python
#author

author = soup.find("a", class_ = "attribution-date__author__link")

author_text = author.text
print(author_text)
```

```python
#images
from IPython.display import Image

images = soup.find_all("img")

for image in images:
    image_url = image.get("src")
    display(Image(url = image_url, unconfined = True))
    print(image_url)
    print(image.get("alt"))
    print("\n")
```

```python
#links

links = soup.find_all("a")

for link in links:
    print(link.get("href"))
    print(link.get_text())
    print("\n")
```
